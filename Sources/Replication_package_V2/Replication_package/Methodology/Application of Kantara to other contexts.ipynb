```python
def chatWithGPT(text):
  completion = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
      {"role": "system", "content": "You are a helpful assistant."},
      {"role": "user", "content": text}
    ]
  )
  print(completion.choices[0].message.content)
  return(completion.choices[0].message.content)
```

```python
import time
d = {'Decision': [], 'Rationale': []}

i = 0

for s in df_oss_commits_5_decision_rationale_sentences['new_message1'][90:]:
#for s in df_button_commits_decision_rationale_sentences['text'][150:200]:
  print(i, s)

  prompt = """
      Which chunks of this sentence indicate Decision and which chunks indicate Rationale?
      Decision is an action or a change that has been made, including a description of the patch behaviour.
      Rationale is the reason for a decision or value judgment.

      Your answer should be strictly in this format:
      Decision: XX
      Rationale: YY
      XX and YY must be chunks from the sentence in question.
      All sentences express both a Decision and the Rationale behind it.
      If you still can not find the decision or the rationale, only answer "None" instead of XX or YY.
      Do not add any other text in your answer.

      Example1:
      s = "Remove the mm_kill() function call because we do not use it anymore"
      Decision: Remove the mm_kill() function call
      Rationale: we do not use it anymore

      Example2:
      s = "Remove the unused mm_kill() function call"
      Decision: Remove the unused mm_kill() function call
      Rationale: unused

      Example3:
      s = "Fix this bug"
      Decision: Fix this bug
      Rationale: Fix

      Sentence :
      """

  answer = chatWithGPT(prompt + s)
  decision = answer.split('\nRationale: ')[0].split('Decision: ')[1]
  rationale = answer.split('\nRationale: ')[1]
  print((decision, 'has rationale', rationale))
  d['Decision'].append(decision)
  d['Rationale'].append(rationale)
  #time.sleep(20)
  i = i+1

df_graph = pd.DataFrame(data=d)
df_graph
```

```python
from sentence_transformers import SentenceTransformer, util

def check_similarity(sentence1, sentence2):
  #model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
  model = SentenceTransformer('distilbert-base-nli-mean-tokens')

  #Compute embedding for both lists
  embedding_1= model.encode(sentence1, convert_to_tensor=True)
  embedding_2 = model.encode(sentence2, convert_to_tensor=True)

  res = util.pytorch_cos_sim(embedding_1, embedding_2)
  print(res)
  return float(res[0][0])
```

```python
# Code to test 1 pair only, not batches
from transformers import pipeline


def detect_contradiction_nli_one_pair(sentence1, sentence2):
    pipe = pipeline( model="roberta-large-mnli", device='cuda:0')

    # Concatenate the sentences and get predictions
    sentence_pairs = str(sentence1 + sentence2)
    predictions = pipe(sentence_pairs)
    print(predictions)

    # Extract prediction +  score
    label = predictions[0]['label']
    score = predictions[0]['score']

    if label == 'CONTRADICTION' and score > 0.5:
      return True, score
    else:
      return False, 0.0
```

## Consistency Mechanism 1

```python
# Does similar decisions have contradictory rationales  ?
from tqdm import tqdm

def consistency_mechanisms_1(graph_dataframe):

  nb_similar = len(graph_dataframe[graph_dataframe['Relationship'] == 'Similar'])

  inconsistencies_dataframe = graph_dataframe.copy()
  inconsistencies_dataframe['similarity_rationales'] = None
  inconsistencies_dataframe['contradiction_rationales'] = None

  for ind in tqdm(graph_dataframe.index):
      if graph_dataframe['Relationship'][ind] == 'Similar':
        s = check_similarity(graph_dataframe['Rationale1'][ind], graph_dataframe['Rationale2'][ind])
        # pred, c = check_entailement(graph_dataframe['Rationale1'][ind], graph_dataframe['Rationale2'][ind])
        contradiction_detected, contradiction_strength = detect_contradiction_nli_one_pair(graph_dataframe['Rationale1'][ind], graph_dataframe['Rationale2'][ind])

        inconsistencies_dataframe['similarity_rationales'][ind] = s
        if contradiction_detected:
          inconsistencies_dataframe['contradiction_rationales'][ind] =  contradiction_strength

  inconsistencies_dataframe.to_csv('similarities_inconsistencies_mechanism1_SPRINGBOOT.csv', sep=';', index=False)
```
