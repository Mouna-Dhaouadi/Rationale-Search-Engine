```python
def chatWithGPT(text):
  completion = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
      {"role": "system", "content": "You are a helpful assistant."},
      {"role": "user", "content": text}
    ]
  )
  print(completion.choices[0].message.content)
  return(completion.choices[0].message.content)
```

#  Prompting

```python
import time
d = {'Decision': [], 'Rationale': []}

i = 0
for s in df_rd['message_preprocessed'][750:775]:
  print(i, s)

  prompt = """
      Which chunks of this sentence indicate Decision and which chunks indicate Rationale?
      Decision is an action or a change that has been made, including a description of the patch behaviour.
      Rationale is the reason for a decision or value judgment.

      Your answer should be strictly in this format:
      Decision: XX
      Rationale: YY
      XX and YY must be chunks from the sentence in question.
      All sentences express both a Decision and the Rationale behind it.
      If you still can not find the decision or the rationale, only answer "None" instead of XX or YY.
      Do not add any other text in your answer.

      Example1:
      s = "Remove the mm_kill() function call because we do not use it anymore"
      Decision: Remove the mm_kill() function call
      Rationale: we do not use it anymore

      Example2:
      s = "Remove the unused mm_kill() function call"
      Decision: Remove the unused mm_kill() function call
      Rationale: unused

      Example3:
      s = "Fix this bug"
      Decision: Fix this bug
      Rationale: Fix

      Sentence :
      """

  answer = chatWithGPT(prompt + s)
  decision = answer.split('\nRationale: ')[0].split('Decision: ')[1]
  rationale = answer.split('\nRationale: ')[1]
  print((decision, 'has rationale', rationale))
  d['Decision'].append(decision)
  d['Rationale'].append(rationale)
  time.sleep(20)
  i = i+1

df_graph = pd.DataFrame(data=d)
df_graph
```
