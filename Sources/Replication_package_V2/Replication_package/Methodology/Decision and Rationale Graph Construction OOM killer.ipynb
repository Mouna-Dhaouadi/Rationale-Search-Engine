```python
#TODO: put device GPU before executing this code
from sentence_transformers import SentenceTransformer, util


def check_similarity(sentence1, sentence2):
  #model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
  model = SentenceTransformer('distilbert-base-nli-mean-tokens')

  # device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
  #model.to(device)
  # print(device)

  #Compute embedding for both lists
  embedding_1= model.encode(sentence1, convert_to_tensor=True)
  embedding_2 = model.encode(sentence2, convert_to_tensor=True)

  res = util.pytorch_cos_sim(embedding_1, embedding_2)
  print(res)
  return float(res[0][0])
```

```python
def verification_mechanisms(graph_dataframe, new_decision):

  # 1. look for similar decisions to this new_decision
  similar_decisions = list (graph_dataframe[(graph_dataframe['Decision1'] == new_decision ) & (graph_dataframe['Relationship'] == 'Similar') ]['Decision2']) \
                     + list(graph_dataframe[(graph_dataframe['Decision2'] == new_decision ) & (graph_dataframe['Relationship'] == 'Similar') ]['Decision1'])

  print('This new decision \n ', new_decision, ' \n is similar with these decisions: \n ', list(similar_decisions))
  print('*****************')

  # 2. Look if there is a contradiction between these similar decisions and any other decisions
  potential_conflicts = []

  for dec in similar_decisions:
    c =  list (graph_dataframe[(graph_dataframe['Decision1'] == dec ) & (graph_dataframe['Relationship'] == 'Contradicts') ]['Decision2'])
    c.extend(list (graph_dataframe[(graph_dataframe['Decision2'] == dec ) & (graph_dataframe['Relationship'] == 'Contradicts') ]['Decision1'])  )
    print('Decision: ', new_decision, ' \n is similar to \n ', dec, ' \n which in turn contradicts these decisions \n', c)
    if len(c) > 0:
      print('SO, possible conflicts \n')
    else:
      print('\n')
    potential_conflicts.extend(c)

  potential_conflicts= list(dict.fromkeys(potential_conflicts))  # remove duplicates in potential_conflicts
  print('******************')
  print('This new deciison \n ', new_decision, ' \n could cause conflicts with these ', len(potential_conflicts) ,' decisions: \n ', potential_conflicts)
```
