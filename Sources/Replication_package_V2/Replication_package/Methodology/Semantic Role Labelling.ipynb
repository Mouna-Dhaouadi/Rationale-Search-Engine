## finer grained analysis, clause level

```python
def srl2(sentence, manner=True, bert=True):

  if bert : predictor=pretrained.load_predictor("structured-prediction-srl-bert")
  else: predictor=pretrained.load_predictor("structured-prediction-srl") # - A reimplementation of a deep BiLSTM sequence prediction model (Stanovsky et al., 2018)
                                                                          # from here: https://docs.allennlp.org/models/main/

  p = predictor.predict(
    sentence= sentence
  )

  PRP = False
  CAU = False
  MNR = False

  verbs={}
  descriptions={}

  nb_verbs = len(p['verbs'])
  print(sentence)
  for i in range(nb_verbs):
    print('verb = ', p['verbs'][i]['verb'],' | description = ', p['verbs'][i]['description'])

    if 'ARGM-PRP' in p['verbs'][i]['description'] : PRP = True ; verbs['PRP']= p['verbs'][i]['verb'] ; descriptions['PRP'] = p['verbs'][i]['description']
    if 'ARGM-CAU' in p['verbs'][i]['description'] : CAU = True ; verbs['CAU']= p['verbs'][i]['verb'] ; descriptions['CAU'] = p['verbs'][i]['description']
    if 'ARGM-MNR' in p['verbs'][i]['description'] : MNR = True ; verbs['MNR']= p['verbs'][i]['verb'] ; descriptions['MNR'] = p['verbs'][i]['description']

  print('PURPOSE? ', PRP, 'CAUSE? ',  CAU,  'MANNER ? ',MNR)


  if manner : return PRP or CAU or MNR, verbs, descriptions
  else: return PRP or CAU, verbs, descriptions
```
